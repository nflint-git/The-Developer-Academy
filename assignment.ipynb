```python
## dice rolling 
import random

repeater = True
while repeater:
    # choice variable to store the user decision
    choice = input("would you like to role the die?")
    
    #if the user types "yes" then print a random die roll
    if choice.lower() == "yes":
        dice_outcome = random.randint(1,6)
        print(dice_outcome)
        

        ## Trying to make the repeater work to roll again
        while True:
            choice_check = input("would you like to roll again?")
            if choice_check.lower() == "yes":  
                dice_outcome = random.randint(1,6)
                print(dice_outcome)
            elif choice_check.lower() == "no":
                repeater = False
                break
            else:
                print("You have not entered correctly.")
                repeater = False
                break
    
    #if type "no"
    elif choice.lower() == "no":
            repeater = False
    
    #if type something else
    else: 
        clarify = input("Would you like to roll the die? Please choose ONLY 'yes' or 'no'.")
        if clarify.lower() == "no":
            repeater = False
        elif clarify.lower() != "yes":
            print("You have not entered correctly")
            repeater = False
        else:
            dice_outcome = random.randint(1,6)
            print(dice_outcome)
            while True:
                choice_check = input("would you like to roll again?")
                if choice_check.lower() == "yes":  
                    dice_outcome = random.randint(1,6)
                    print(dice_outcome)
                elif choice_check.lower() == "no":
                    repeater = False
                    break
                else:
                    print("You have not entered correctly.")
                    repeater = False
                    break
```

```python
## dice rolling 

# this uses the random library and the randint method 
## to choose a random integer that includes the full specified range i.e. 1 to 6
import random 

#create an infinite loop to allow for multipe iterative rolls of the die
loop = True
while loop == True:

    # this is a variable to store the user's initial decision
    choice = input("Would you like to roll a 6 sided die? (Y/N)")

    # this is a variable for the outcome of the roll 
    die_outcome = random.randint(1,6)

    #create the first outcome being yes
    #use the .lower method to remove any differences in capitalisation input 
    if choice.lower() == "y":
        print(die_outcome)
        
        #indefinite loop for repeated rolls after a first succesful roll
        while True:
            #the decisions for repeated rolls
            again = input("Would you like to roll again? (Y/N)")
            if again.lower() == "y":
                die_outcome = random.randint(1,6) #to reroll the die
                print(die_outcome)
            elif again.lower() == "n":
                loop = False #this ends the outer while loop
                break        #this ends the inner/ nested while loop
            else:
                if again.lower() != "y" or "n": #the input is misspecified
                    print("You have entered incorrectly, please try again.")
                continue #allows for another attempt to input

    #create the second outcome where the user has said no
    elif choice.lower() == "n":
        loop = False #this ends the loop
    
    #the third outcome where the user has inputted incorrectly
    else:
        if choice.lower() != "y" or "n":
            print("You have entered incorrectly, please try again.")
        continue


## Note: for some reason, sometimes the process skips an output. 
## If it does the next option has two outputs. 
## Not sure why. More likely to happen over extended rolls.
```
